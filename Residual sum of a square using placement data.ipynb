```python
import pandas as pd
```

```python
dataset=pd.read_csv("preplacement.csv")
dataset
```

```python
dataset.drop(["sl_no","Unnamed: 0"],inplace=True,axis=1)
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p','gender_M',
       'ssc_b_Others', 'hsc_b_Others', 'hsc_s_Commerce', 'hsc_s_Science',
       'degree_t_Others', 'degree_t_Sci&Tech', 'workex_Yes',
       'specialisation_Mkt&HR', 'status_Placed']]
independent
```

```python
dependent=dataset[['salary']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_test
```

```python
from sklearn.svm import SVR
```

```python
regressor=SVR(kernel="rbf",C=3000)
```

```python
regressor.fit(x_train,y_train)
```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python
regressor.support_
```

```python
y_pred=regressor.predict(x_test)
y_pred=y_pred.reshape(65,1)
```

```python
y_test
```

```python
residuals = y_test - y_pred
residuals
```

```python
import numpy as np
rss = np.sum(residuals**2)
rss
```

```python
print("Residual Sum of Squares (RSS):", rss)
```
